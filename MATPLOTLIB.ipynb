## **MATPLOTLIB**

Imagem1.png

```python
import matplotlib.pyplot as plt
```

PYPLOT - submodulo da biblioteca matplotlib

**GRAFICO DE LINHA**

```python
#plot()
#motando o grafico básico
x = [1,2,3,4]
y = [2,3,4,3]

plt.plot(x,y)
plt.show() #apresenta o grafico
```

```python
#adicionar uma legenda ao grafico
plt.plot(x,y, label='dados')
#legenda para o eixo y
plt.ylabel('eixo y')
#legenda para o eixo x
plt.xlabel('eixo x')
#titulo do grafico
plt.title('titulo do grafico')
#escala do x
plt.xticks([1,2,3,4])
#escala do y
plt.yticks([1,2,3,4])
#legenda do grafico, que foi incluido la em cima
plt.legend()
plt.show()
```

**LINES**

```python
#mudança no grafico do grafico
#linestyle - muda o estilo da linha
#color - cor da linha
#lw - largura da linha
plt.plot (x,y, linestyle='dashdot', color='red', lw=3.0)
plt.show()
```

**GRAFICO DE DISPERSÃO**

.SCATTER()

```python
plt.scatter(x,y, color='k')
plt.show()
```

**GRAFICO DE BARRA**

.BAR()

```python
plt.bar(x,y, color='yellow')
```

**EIXOS**

Mudando os tipo de marcadores do grafico

```python
#grafico de dispersao, mudando o marcador
plt.scatter(x,y, color='k', marker = 'v')
```

Alterando os limites do grafico

```python
plt.plot(x,y)
plt.axis(xmin=-1, xmax=10, ymin=0, ymax=12)
plt.show()
```

**SUBPLOTS**

```python
valoresx = [1,2,3,4]
valoresy = [1,4,2,3]

#figura - armazena coisas graficas
figura = plt.figure(figsize=(20,3))
figura.suptitle('Titulo geral')

figura.add_subplot(131) #1 uma linha, 3 colunas, 1 - posição 1
plt.plot(valoresx, valoresy, label='Um dado qualquer')
plt.legend() #para aparecer a legenda
plt.title('Gráfico 1')

figura.add_subplot(133)
plt.scatter(valoresx, valoresy, color='red')
plt.title('Grafico 2')

figura.add_subplot(132)
plt.bar(valoresx, valoresy, color='yellow')
plt.title('Grafico 3')

plt.savefig('grafico.png')

plt.show(figura)
```
