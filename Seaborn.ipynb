Dados estatisticos, ele roda uma camada acima do matplotlib


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_videos = pd.read_excel('videosVT.xlsx')
display(df_videos)
```

  Dispersão

```python
#hue = 'Responsavel' - ele da cor para cada responsavel
fig = sns.scatterplot(data=df_videos, x='Nº de View', y='Nº de Like', hue="Categoria", style="Responsável")
plt.show(fig)
```

```python
fig = sns.scatterplot(data=df_videos, x='Nº de View', y='Nº de Like', hue="Responsável", palette=['red','black'])
plt.show(fig)
```

Dispersão relacional

```python
graf_rel = sns.relplot(data=df_videos, x='Nº de View', y='Nº de Like', hue='Categoria', col='Responsável')
graf_rel.set_titles('O Responsável deste gráfico: {col_name}')
plt.show(graf_rel)
```
